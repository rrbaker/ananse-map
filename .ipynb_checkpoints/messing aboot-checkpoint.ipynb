```python
import pandas as pd
import re
from numpy import nan
from collections import defaultdict
```

```python
df = pd.read_excel("new_spaces_aliases.xlsx")
```

```python
raw_websites = df.Website
processed_websites = []
domainname = re.compile("^((?!-))(xn--)?[a-z0-9][a-z0-9-_]{0,61}[a-z0-9]{0,1}\.(xn--)?([a-z0-9\-]{1,61}|[a-z0-9-]{1,30}\.[a-z]{2,})")
rejected = set()
website_index = defaultdict(list)
website_strings = []
bonus_sites = []
for i, string in enumerate(raw_websites):
    if string is nan:
        continue
    try:
        websites = string.split()
    except:
        print "ERROR: ", string
        continue
    parsed_sites = []
    for website in websites:
        website = website.lower()
        if website[0] in ".<":
            website = website[1:]
        for prefix in ["site=", "http://", "https://", "www.", "www3."]:
            website = re.sub(".*"+prefix, "", website)
        if website[-1] in ">/":
            website = website[:-1]
#         website = "".join(re.split("[^a-z0-9/.-]", website))
        if not domainname.search(website):
#             print "rejected:", website
            rejected.add(i)
            continue
#         print "out:", website
        website_index[website].append(i)
        parsed_sites.append(website)
    if len(parsed_sites) > 1:
        print i, parsed_sites
    website_strings.append(", ".join(parsed_sites))
print "rejected rows", len(rejected)
```
